# Central Limit Theorem Visualized

Write a program that illustrates, in some way, the Central Limit Theorem (preferably with one or more figures)

   - For example, the Central Limit Theorem indicates that the normalized sum (i.e. the arithmetic mean) of random variables, each independently sampled from the same distribution, will asymptotically follow a Gaussian distribution. Try sampling a continuous variable from a probability distribution that is *not* a Gaussian - repeat this N times and take the average. Then repeat *that* again many times (so M "experiments" each with N samples) - what does this distribution of averages look like? As you increase N?

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#We will highlight the central limit theorem with a coin flip example. 

def Sampler(rate):
    return stats.bernoulli.rvs(rate)

def Frequency(Infile):
    
    InputFile = Infile
    Nroll = 1
    Npass0 = []
    Npass_min = 1e8
    Npass_max = -1e8

    with open(InputFile) as ifile: #Opens files
        for line in ifile: #
            lineVals = line.split() #splits lines ups
            Nroll = len(lineVals) #obtains the number of rolls
            Npass = 0 #defines the number of experiments
            for v in lineVals: #Chance of getting a 1 i.e the house winining the game. 
                Npass += float(v)

            if Npass < Npass_min:
                Npass_min = Npass
            if Npass > Npass_max:
                Npass_max = Npass 
            Npass0.append(Npass)
            
    return (Npass0,Nroll)
```

```python
#The actual Experiment
Nexp0 = 100
Nexp1 = Nexp0*10
Nexp2 = Nexp1*10
Roll = 100
rate = 0.5

outfile = open("Rolls_Central_100.txt", 'w+')
for e in range(0,Nexp0):
    for t in range(0,Roll):
        outfile.write(str(Sampler(rate)) +" ")
    outfile.write(' \n')
outfile.close()

outfile = open("Rolls_Central_1000.txt", 'w+')
for e in range(0,Nexp1):
    for t in range(0,Roll):
        outfile.write(str(Sampler(rate)) +" ")
    outfile.write(' \n')
outfile.close()

outfile = open("Rolls_Central_10000.txt", 'w+')
for e in range(0,Nexp2):
    for t in range(0,Roll):
        outfile.write(str(Sampler(rate)) +" ")
    outfile.write(' \n')
outfile.close()
```

```python
#Data file name
InputFile0 = "Rolls_Central_100.txt"
InputFile1 = "Rolls_Central_1000.txt"
InputFile2 = "Rolls_Central_10000.txt"
```

```python
#frequency of outputs
Output100   = Frequency(InputFile0)
Output1000  = Frequency(InputFile1)
Output10000 = Frequency(InputFile2)
```

```python
print(Output100)
print("\n  ")
print(Output1000)
print("\n")
print(Output10000)
```

```python
title = str(Roll) +  " rolls / " + str(Nexp0) +" experiment"

# make Npass figure
fig, ax = plt.subplots() 
plt.title(title)
plt.legend("sksj")
plt.xlabel("Frequency")
plt.ylabel("Probability")
hist1 = ax.hist(Output100[0], Output100[1], density=True, color='b', alpha=0.7)
```

```python
title = str(Roll) +  " rolls / " + str(Nexp1) +" experiment"

# make Npass figure
fig, ax = plt.subplots() 
plt.title(title)
plt.legend("sksj")
plt.xlabel("Frequency")
plt.ylabel("Probability")
hist2 = ax.hist(Output1000[0], Output1000[1], density=True, color='b', alpha=0.7)
```

```python
title = str(Roll) +  " rolls / " + str(Nexp2) +" experiment"

# make Npass figure
fig, ax = plt.subplots() 
plt.title(title)
plt.legend("sksj")
plt.xlabel("Frequency")
plt.ylabel("Probability")
hist3 = ax.hist(Output10000[0], Output10000[1], density=True, color='b', alpha=0.7)
```

# Discussion:

As you can see as we increase the number of experiments, the distribution tends to a gaussian. This is to be expected according to the central limit theorem. 
If we were to incread the number of measurments, this would simply make the distributions more "refined" meaning, we would be more concentrated around specific freqeuncies.
